## <span style = 'background :lightblue'>STEP 3 : Down-sampling of Train and Test DataSet</span>

```python
import copy
```

```python
from imblearn.under_sampling import RandomUnderSampler, NearMiss, EditedNearestNeighbours, RepeatedEditedNearestNeighbours, CondensedNearestNeighbour, OneSidedSelection, NeighbourhoodCleaningRule, InstanceHardnessThreshold
```

```python
under_sampl_models = [RandomUnderSampler(), 
                      NearMiss(), 
                      EditedNearestNeighbours(),
                      CondensedNearestNeighbour()]
```

```python
def undersampling(us_model = None ,sampl_stra_train = 1, sampl_stra_test = 1,  
                  Xtrain = None , ytrain = None, 
                  Xtest = None, ytest = None ,
                  rand_state = 100):

    # On Training Datatset
    model_train = copy.deepcopy(us_model)

    if str(model_train) in [str(RandomUnderSampler()),str(NearMiss())]:

        model_train = model_train.set_params(**{"sampling_strategy" : sampl_stra_train 
                                               })
                                                                                  
    else :
        
        model_train = model_train.set_params(**{"sampling_strategy" : "majority"})
                                            
    # fit and apply the transform
    X_train_res, Y_train_res = model_train.fit_resample(Xtrain, ytrain)


    # On Test Dataset
    model_test = copy.deepcopy(us_model)

    if str(model_test) in [str(RandomUnderSampler()),str(NearMiss())]:

        model_test = model_test.set_params(**{"sampling_strategy" : sampl_stra_test , 
                                             }) 
                                            
    else :
        
        model_test = model_test.set_params(**{"sampling_strategy" : "majority"})
                                           
    # fit and apply the transform
    X_test_res, Y_test_res = model_test.fit_resample(Xtest, ytest)

    
    print("\n----------------------------------------------------")
    print(f"\33[1mUnder Sampling Model : {str(us_model)}\33[0m")
    print("----------------------------------------------------")
    print("\n")
    print("X_train_res shape : ",X_train_res.shape)
    print("y_train_res shape : ",Y_train_res.shape)
    print("y_train_res class count : \n" ,Y_train_res.value_counts(),"\n")

    print("X_test_res shape : " , X_test_res.shape)
    print("y_test_res shape : " , Y_test_res.shape)
    print("y_test_res class count : \n", Y_test_res.value_counts(),"\n")


    return (X_train_res, Y_train_res , X_test_res , Y_test_res)
```

## <span style = 'background :lightblue'>STEP 4 : Performance of ML Models on Different Under-Sampled Train and Test DataSets</span>

```python
from sklearn.metrics import classification_report , accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
models = [LogisticRegression(max_iter=1000), SVC(), GaussianNB(),
          DecisionTreeClassifier(), RandomForestClassifier()]
```

```python
for samp_model in under_sampl_models:
    
    
    X_train_res, y_train_res , X_test_res , y_test_res = undersampling(us_model = samp_model ,
                                                                        sampl_stra_train = 0.928, 
                                                                        sampl_stra_test = 0.854,  
                                                                        Xtrain = X_train , 
                                                                        ytrain = y_train , 
                                                                        Xtest = X_test,
                                                                        ytest = y_test ,
                                                                        rand_state = 100)

    for model in models :

        model.fit(X_train_res.values, y_train_res)
        y_hat_train = model.predict(X_train_res.values)
        y_hat_test =  model.predict(X_test_res.values)

        print(f"******************* \33[1m{str(model)}\33[0m *******************\n")

        print("\33[1mTraining Accuracy :\33[0m \n")
        print(accuracy_score(y_train_res,y_hat_train ))

        print() 

        print("\33[1mTesting Accuracy :\33[0m \n")
        print(accuracy_score(y_test_res,y_hat_test ))

        print()
```

### Conclusion :

- Best Train and Test performance is obtained by **EditedNearestNeighbours** Undersampling algorithm.
